### 5. Escrita de Arquivos CSV:

```python
dados_modificados = pd.read_csv("dados_modificados")

dados_modificados
```

## Parte 3 - Manipulação de Dados

### 6. Selecionando Dados:

```python
dm = pd.DataFrame(dados_modificados)

dm.head(10)
```

```python
col_selec = dm[['Name', 'Sex', 	'Age']]

col_selec.head()
```

### 7. Filtragem de Dados:

```python
filtroIdade = dm[dm['Age'] > 30]

filtroIdade
```

```python
filtroFemSobr = dm[(dm['Sex'] == 'female') & (dm['Survived'] == 1)]

filtroFemSobr
```

### 8. Operações de Agregação:

```python
agrupPass = dm.groupby('Pclass')['Age'].mean()

agrupPass
```

```python
agrupPassCont = dm.groupby('Pclass')['Embarked'].count()

agrupPassCont
```

### 9. Criação de Novas Colunas:

```python
dm['Idade_10_anos'] = dm['Age'] + 10

dm
```

## Parte 4 - Desafios Extras

### 10. Removendo Valores Nulos:

```python
mediaIdade = dm['Age'].mean()

valoresNulosSubs = dm.fillna(mediaIdade)

valoresNulosSubs
```

### 11. Ordenação de Dados:

```python
dmOrdenado = dm.sort_values('Age')

dmOrdenado
```

### 12. Exportação para Excel:

```python
dm.to_excel("titanic_tratado.xlsx") 
```
